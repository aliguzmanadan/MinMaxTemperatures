<h1>3. Polynomial regression <a class="anchor" id="Section3"></a>

In this section, we look for the best polynomial model that fits our data. To that end, we will use a Ridge model which considers a regularization parameter proportional to the L2 norm. See https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html

We now define a function that does a grid search along the regularization parameters alpha and the degree of the polynomial. This returns

- A dictionary with the CV average $R^2$ errors for each pair of hyperparameters (degree, alpha).
- A dictionary with the standar error (variance of the mean) of each set of individual cv scores corresponding to each pair of hyperparameters (degree, alpha).
- The best pair (degree, alpha) yielding the best estimator based on the CV.



```python
def GridSearchPolRegCV(list_degrees, list_alphas, X, Y, cv = 5, random_state = 47):
    
    cv_avgScores = {}
    cv_stErrors = {}
    
    grid_params = [(d,a) for d in list_degrees for a in list_alphas]
    
    #For each combination of degree and alpha compute the average Rˆ2 CV score
    for d, a in grid_params:
        poly = PolynomialFeatures(degree=d)
        poly_feats = poly.fit_transform(X)
        model = Ridge(alpha = a, random_state = random_state)
        scores_r2 = cross_val_score(model, poly_feats, Y, scoring = 'r2', cv=cv)
        cv_avgScores[(d,a)] = scores_r2.mean()
        cv_stErrors[(d,a)] = scores_r2.std()/math.sqrt(cv)
        
    
    best_params = max(cv_avgScores, key = cv_avgScores.get)
    best_score = cv_avgScores[best_params]
        
    return cv_avgScores, cv_stErrors, best_params, best_score, 
```

We now do a grid search for polynomials of degree 1, ... 20.

```python
degrees = list(range(1,21))
alphas = [0] + [0.1*pow(2,i) for i in range(30)]

#Amount of fits is len(degrees)*len(alphas)*cv
print(f'Fitting {len(degrees)*len(alphas)*10} models')


import time
start_time = time.time()
scores, errors, best_params, best_score = GridSearchPolRegCV(degrees, alphas, X_train, Y_train, cv=10)
print(f'The best parameters are {best_params} with and average R^2 error of {best_score}.')
print(f'CV Execution time: {time.time() - start_time} seconds')
```

Let's see how the best model found performs in the test set. 

```python
#Extract hyperparametets 
random_state = 47
degree = best_params[0]
alpha = best_params[1]

pf = PolynomialFeatures(degree= degree)
X_trans = pf.fit_transform(X_test)
model = Ridge(alpha = alpha, random_state = random_state)
model.fit(X_trans, Y_test)
model.score(X_trans, Y_test)
```

Let us now plot this estimated curve and visually check how well it fits the data. This sould be kind of a overfitting situation.

```python


#Fir model on train set
poly = PolynomialFeatures(degree=degree)
poly_feats = poly.fit_transform(X_train)
model = Ridge(alpha = alpha, random_state = random_state)
model.fit(poly_feats, Y_train)

#Predict and Plot model
X_new = np.linspace(min(X)-0.3, max(X)+0.3, 2000).reshape(2000,1)
pf = PolynomialFeatures(degree= degree)
X_trans = pf.fit_transform(X_new)
Yhat_trans = model.predict(X_trans)

#Plotting predictions
width = 30
height = 15
plt.figure(figsize=(width, height))

plt.scatter(X,Y)
plt.plot(X_new, Yhat_trans, color='red', linewidth=4)
plt.title('MinTemp vs MaxTemp')  
plt.xlabel('MinTemp')  
plt.ylabel('MaxTemp')
plt.show()
```

The R^2 score is a bit better than plane linear regression. Also, the picture seems good at first.

But the real question is: Is it good enough? Does this improvement in R^2 is enough to justify the increasing in variance? Maybe there are other simpler models that are just as good.

<h1>4. Model comparison <a class="anchor" id="Section4"></a>

Let's plot the $R^2$ scores obtained for each degree

```python
avg_scores = [scores[i] for i in scores]
degrees = [i[0] for i in scores]

fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(degrees, avg_scores, c='blue', label='data')
plt.axis('tight')
plt.title("Hyperparameter selection")
ax.set_xlabel("degree");
ax.set_ylabel("CV R^2");
plt.tight_layout()
plt.show()
```

This is weird! The plot looks bad: there are very negative R2's, and it's so big that the interesting bit of the plot is completely compressed. 

We see that the smallest negative values are obatained for higher degree polynomials. This should be due to the high variance, spacially in those cases where the regularization coefficient is very small. These are of course the highest variance scenarios.  

Let us check that indeed these cases correspond, to high degree and low alphas.


```python
bad_params = [(d,a) for (d,a) in scores if scores[(d,a)] < -1]
bad_params
```


Let's now get rid of this set of bad parameters and plot again our comparison.

```python
avg_scores = [scores[i] for i in scores if scores[i] > 0]
degrees = [i[0] for i in scores if scores[i] > 0]

fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(degrees, avg_scores, c='blue', label='data')
plt.axis('tight')
plt.title("Hyperparameter selection")
ax.set_xlabel("degree");
ax.set_ylabel("CV R^2");
plt.tight_layout()
plt.show()
```

This is a much better picture! We see that removing the high variance cases paid off. Of course, now we still have high bias models corresponding to very high values of the regularization coefficient alpha. These should be the lower points in this plot. So we will get rid of the lower $R^2$ scores to have a better view of the top of the plot which is now compressed.

We also see that from degrees 5 on there seems to be not much improvement. There some asimptotic behaivor in our plot.

We shall select our final model ussing an error bar. This is based on the standard error (std) of the cv average socores with respect to their mean. 

The idea is to take an horizontal error bar whose value in the R^2 axis is given by the highest  average R<sup>2</sup> minus the std of the errors. Then one chooses the lowest degree / highest alpha (least complicated model) that has its average R<sup>2</sup> above that bar.

For this we assume that the possible noise error in the CV averages is normally distributed N(0,1). So, our error bar is set to $highestCvScore - 1.96*sts_error$, where $sts_error$ is the standard error of the CV scores corresponding to the best hyperparameters.

We also recall that 1.96 is the approximate value of the 97.5 percentile point of the standard normal distribution. This means that 95% of the area under a normal curve lies within roughly 1.96 standard deviations of the mean, and due to the central limit theorem, this number is therefore used in the construction of approximate 95% confidence intervals.


```python
avg_scores = [scores[i] for i in scores if scores[i] > 0.7]
degrees = [i[0] for i in scores if scores[i] > 0.7]


fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(degrees, avg_scores, c='blue', label='data')
#Drawwing the error bar 
plt.axhline(y=best_score-1.96*errors[best_params])
plt.axis('tight')
plt.title("Hyperparameter selection")
ax.set_xlabel("degree");
ax.set_ylabel("CV R^2");
plt.tight_layout()
plt.show()
```

So, although the most optimal tuple of hyperparameters is (18, 819.2), there us not  a lot of evidence that orders 5 and 6 are really worse. Let's dive in and get the simplest model scoring above the bar.

```python
bar = best_score-1.96*errors[best_params]

good_params = [i for i in scores if scores[i] > bar]
good_params
```

The most simple model is that of lowest degree and of highest regularization parameter possible.

```python
simplest = min(good_params, key = lambda i: (i[0], -i[1]))
simplest
```

We now compare the optimal model found with this simple model that still is good enoguh.

```python
#Parameters for the best model found and also for the simplest model found that performs good enough
deg_best = best_params[0]
al_best = best_params[1]
deg_s = simplest[0]
al_s = simplest[0]

#Train models
reg_simple = Pipeline([('poly', PolynomialFeatures(degree=deg_s)),
                  ('ridge', Ridge(alpha = al_s, random_state = random_state))])
reg_best = Pipeline([('poly', PolynomialFeatures(degree=deg_best)),
                  ('ridge', Ridge(alpha = al_best, random_state = random_state))])
reg_simple.fit(X_train, Y_train)
reg_best.fit(X_train, Y_train)

#predictions
X_plot = np.linspace(min(X)-0.3, max(X)+0.3, 2000).reshape(2000,1)
Ysimple_plot = reg_simple.predict(X_plot)
Ybest_plot = reg_best.predict(X_plot)

#Plot the fit
width = 30
height = 15
plt.figure(figsize=(width, height))

plt.scatter(X,Y, color='grey')
plt.plot(X_plot, Ysimple_plot, color='blue', linewidth=4, label=f'degree {deg_s}')
plt.plot(X_plot, Ybest_plot, color='red', linewidth=4, label=f'degree {deg_best}')
plt.title('MinTemp vs MaxTemp')  
plt.xlabel('MinTemp')  
plt.ylabel('MaxTemp')
plt.legend()
plt.show()
```

We see that our simple model take less bends (less variance) that the optimal one, and still performs similarly on new data. 

Let's check that it also desibes the data distribtuion better.

```python
#Predictions on test set
Ysimple_test = reg_simple.predict(X_test)
Ybest_test = reg_best.predict(X_test)

#Rˆ2 scores
r2_simple = reg_simple.score(X_test, Y_test)
r2_best = reg_best.score(X_test, Y_test)
print(f'R^2 score of simple model: {r2_simple}')
print(f'R^2 score of oprimal model: {r2_best}')   


#Ploting distributions
width = 30
height = 15
plt.figure(figsize=(width, height))


ax1 = sns.distplot(Y_test, hist=False, color="grey", label="Actual distribution")
sns.distplot(Yhat_test, hist=False, color="green", label="Lniear model distribution" , ax=ax1)
sns.distplot(Ysimple_test, hist=False, color="b", label="Distribition simple model" , ax=ax1)
sns.distplot(Ybest_test, hist=False, color="r", label="Distribition optimal model" , ax=ax1)

plt.title('Distribution')  
plt.xlabel('MaxTemp')
plt.ylabel('Density of observations')

plt.show()
plt.close()
```

<h1>5. Concluding remarks <a class="anchor" id="Section5"></a>

We see that our simple model (blue) describes better the distribution of the data than the optimal model (red). The latter yields the same mode of the data and a somewhat similar distribution as the original. The main difference is that the frequency of the mode is higher for our model than in reality. 

In comparison, the linear model studied at the begining of the notebook has a similar frequency for the mode but the distribution graph is shifted to the right. As a consequence, it predicted a higher mode than originally. 

All in all, the simple polynomial model seems to perform good enough, when compared with the very optimal and with the linear model. 

**So we conclude that a 5 degree polynomial, with tha high regularizartion parameter, performs good enough. It manages to explain the variability of new seen data with around of 80% of accuracy (i.e. 0.80 $R^2$ score).**

This performance could be improved by using algorithms that localize better the fit. Such algotrithms include:
* Decision tree type of models
* Splines

Another possibility could be to look for ourliers in the data and remove them. There are some min temperatures to which correspond a wide range of max temperatures. Looking into these range of tempretatures and removing possible outliers could help to improve the predictions. 

Nevertheless, this data set seems to allow for a good enough description by meand of linear regression.
